Un cierto sensor se deja abierto por un segundo. El sensor consiste en $r$ pixeles y en promedio el sensor es golpeado por $2$ partículas por segundo, tal que la probabilidad de detectar una partícula en un pixel dado es de $2/r$ (para $r \gg 2$). Grafique la distribución de probabilidad de observar $n$ partículas para un pixel dado, en función de $n$, para diferentes valores de $r$:

* Usando la distribución binomial.

* ¿Cuándo es posible usar la distribución de Poisson?

### Datos

La fórmula para la distribución binomial es:
$$
P(n;r,p) = \frac{r!}{n!(n-r)} p^{n}(1-p)^{r-n}
$$

Tenemos:

- Probabilidad de detectar una particula en un pixel = 2/r (p)
- Éxitos: observar n partículas (n)
- r son los píxeles a del sensor (a elegir), lo que se puede interpretar como la cantidad de veces a realizar el experimento.



```python
# Importamos módulo para la realizar combinatoria en dist. binomial.
from scipy.special import comb

def binomial(n_particulas, r_pixels):
    '''
    Función para realizar distribución binomial. Solicita como argumento la cantidad de partículas que se 
    desea observar y los pixeles, la probabilidad depende del número de pixeles por lo tanto no se requiere.
    '''
    p = 2/r_pixels
    # Como p está definido por defecto, no es necesario solicitarlo en la función binomial.
    return comb(r_pixels, n_particulas)*p**n_particulas*(1-p)**(r_pixels - n_particulas)
```

```python
binomial(5, 20) # Un ejemplo.
```

```python
# Importamos módulo para graficar
import matplotlib.pyplot as plt
# Preparamos los datos, la función se encarga de generarlos para distintos valores de n.
def probs(r_pixels):
    '''
    Función que retorna una lista con datos de distintas distribuciones binomiales para
    distintos valores de n.
    '''
    return [binomial(n_particulas, r_pixels) for n_particulas in range(r_pixels+1)]
```

```python
# Generamos el gráfico, se generarán 6 gráficos con r aumentando de valor.

fig = plt.figure(figsize=(8,20), dpi= 100)

ax1 = fig.add_subplot(8,2,1)
ax2 = fig.add_subplot(8,2,2)
ax3 = fig.add_subplot(8,2,3)
ax4 = fig.add_subplot(8,2,4)
ax5 = fig.add_subplot(8,2,5)
ax6 = fig.add_subplot(8,2,6)



ax1.bar(range(5+1), probs(5))
ax2.bar(range(15+1), probs(15))
ax3.bar(range(25+1), probs(25))
ax4.bar(range(50+1), probs(50))
ax5.bar(range(70+1), probs(70))
ax6.bar(range(100+1), probs(100))

        
ax1.set_title('r = 4, p = 0.5')
ax2.set_title('r = 5, p = 0.4')
ax3.set_title('r = 15, p = 0.13')
ax4.set_title('r = 25, p = 0.08')
ax5.set_title('r = 40, p = 0.05')
ax6.set_title('r = 50, p = 0.04')


fig.tight_layout(pad=2.0)


fig.show()
```

¿Cuándo es posible usar la distribución de Poisson?

La distribución de Poisson se puede utilizar cuando r tiende a infinito, por lo tanto, como en este caso podemos hacer variar el valor de los pixeles y es un valor grande para el contexto (pixeles en un sensor), podríamos usar esta distribución.
